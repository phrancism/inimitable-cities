```python
import os
import re
```

```python
CITIES_DIR_PATH = "cities"
CITIES_JSON_DIR_PATH = f"{CITIES_DIR_PATH}/json"
CITIES_UNIQUE_WORDS_JSON_FILE_NAME = "cities_unique_words.json"
CITIES_WORDS_JSON_FILE_NAME = "cities_words.json"
CHARACTERS_TO_REMOVE_REGEX = r"[\.\,\"\(\)\:\;\!\?\-\“\”]"
```

```python
cities_file_names = [item for item in os.listdir(CITIES_DIR_PATH) if os.path.splitext(item)[1] == ".txt"]
```

```python
cities_text = {}
```

```python
for city_file_name in cities_file_names:
    with io.open(f"{CITIES_DIR_PATH}/{city_file_name}", encoding='utf-8') as city_file:
        city_name = city_file_name[:-4] # Remove '.txt'
        cities_text[city_name] = (
            re.sub(CHARACTERS_TO_REMOVE_REGEX, "", city_file.read())
            .replace("'", " ")
            .replace("’", " ")
            .replace("–", " ")
            .replace("  ", " ")
            .lower()
            .strip()
        )
```

```python
# JSON for all words for each city
import json
with io.open(f"{CITIES_JSON_DIR_PATH}/{CITIES_WORDS_JSON_FILE_NAME}", 'w', encoding='utf-8') as city_file:
    cities_words_json = [{"name": city, "words": words.split(' ')} for (city, words) in cities_text.items()]
    json.dump(cities_words_json, city_file, ensure_ascii=False, indent=4)
```

```python
cities_words = {city: text.split(' ') for (city, text) in cities_text.items()}
```

```python
# Unique within itself
unique_words_in_each_city = {city: set(text.split(' ')) for (city, text) in cities_text.items()}
```

```python
unique_words_per_city = [] # Unique compared to other cities
for city, words in unique_words_in_each_city.items():
    city_unique_words = {"name": city, "unique_words": []}
    other_cities_unique_words_copy = unique_words_in_each_city.copy()
    other_cities_unique_words_copy.pop(city)
    other_cities_words_list = list(map(lambda x: list(x), other_cities_unique_words_copy.values()))
    other_cities_words_list_flattened = sum(other_cities_words_list, [])
    # print(other_cities_words_list_flattened)
    for word in cities_words[city]:
        if word not in other_cities_words_list_flattened:
            try:
                if word not in city_unique_words["unique_words"]: city_unique_words["unique_words"].append(word)
            except:
                city_unique_words["name"] = []
    unique_words_per_city.append(city_unique_words)
```

```python
# JSON for all unique words for each city [{city_name:,[unique_words]}]
import json
with io.open(f"{CITIES_JSON_DIR_PATH}/{CITIES_UNIQUE_WORDS_JSON_FILE_NAME}", 'w', encoding='utf-8') as city_file:
    json.dump(unique_words_per_city, city_file, ensure_ascii=False, indent=4)
```

```python
# JSON for all unique words for each city [{"name": string, "unique_words": [string]}]
new_json = []
with io.open(f"{CITIES_DIR_PATH}/{CITIES_UNIQUE_WORDS_JSON_FILE_NAME}", 'r', encoding='utf-8') as city_file:
    data = json.load(city_file)
    new_json = [{"name": city, "unique_words": words} for (city, words) in data.items()]
```
